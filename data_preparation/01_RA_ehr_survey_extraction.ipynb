## RA_case

```python
ra_case_person_df = pd.read_csv('ra_case_person_df.csv')
ra_case_person_df.head()
```

```python
ra_case_survey_df = pd.read_csv('ra_case_survey_df.csv')
ra_case_survey_df.head()
```

```python
ra_survey_info = ra_case_survey_df.loc[(ra_case_survey_df['question']=='Are you still seeing a doctor or health care provider for rheumatoid arthritis (RA)?')]
ra_survey_info.shape
```

```python
ra_case = ra_case_person_df.merge(ra_survey_info[['person_id', 'survey_datetime']], on = 'person_id', how = 'inner')
ra_case.head()
```

```python
ra_case.shape
```

```python
ra_case_condition_df = pd.read_csv('ra_case_condition_df.csv')
ra_case_condition_df.head()
```

```python
ra_case_condition_df.shape
```

```python
# datetime for 1st code of RA
ra_condition = ra_case_condition_df.loc[ra_case_condition_df.groupby('person_id')['condition_start_datetime'].idxmin()]
person_id_with_ra_code = list(set(ra_case_condition_df['person_id']))
len(person_id_with_ra_code)
```

```python
ra_case_ehr_code.head()
```

```python
# last EHR datetime
ra_case_ehr_code = ra_case_ehr_code.loc[ra_case_ehr_code.groupby('person_id')['condition_start_datetime'].idxmax()]
ra_case_ehr_code.shape
```

```python
# merge proxy time data.
ra_case_ehr_code = ra_case_ehr_code[['person_id', 'condition_start_datetime']].merge(ra_condition[['person_id', 'condition_start_datetime']], on='person_id', how='left', suffixes=('_ehr', '_ra'))
ra_case_ehr_code.head()
```

```python
ra_case = ra_case.merge(ra_case_ehr_code[['person_id', 'condition_start_datetime', 'condition_start_datetime_ehr', 'ra_code_YN']], on = 'person_id', how = 'left')
ra_case.shape
```

```python
# Create ra_survey column (1 if asnwer yes in RA_survey, 0 if answer no, NA if missing)
ra_case['ra_event_survey'] = np.where(ra_case['survey_datetime'].notna(), 1, None)
ra_case.head(20)
```

## Control

```python
ra_control_person_df = pd.read_csv('ra_control_person_df.csv')
ra_control_person_df.head()
```

```python
ra_control_person_df.shape
```

```python
ra_control_id = ra_control_person_df['person_id']
```

```python
ra_control_by_survey_survey_df = pd.read_csv('ra_control_by_survey_survey_df.csv')
ra_control_by_survey_survey_df.head()
```

```python
ra_control_by_survey_survey_df.shape
```

```python
# a participant may provide multiple answers at the same datetime.
ra_control_by_survey_survey_df = ra_control_by_survey_survey_df.drop_duplicates(subset='person_id', keep='first')
ra_control_by_survey_survey_df.shape
```

```python
ra_control_by_survey_survey_df['answer'].value_counts()
```

```python
# person with datetime answered "No" in RA survey, with NaT datetime didn't take the survey.
ra_control_person_df = ra_control_person_df.merge(ra_control_by_survey_survey_df[["person_id", "survey_datetime"]], how = "left")
```

```python
ra_control_person_df.head()
```

```python
import pandas as pd
ra_control_ehr_code_sql = """
    SELECT
        person_id,
        condition_concept_id,
        condition_start_datetime
    FROM
        `""" + os.environ["WORKSPACE_CDR"] + f""".condition_occurrence` 
    WHERE person_id in {tuple(ra_control_id)}
    """
ra_control_ehr_code = pd.read_gbq(
    ra_control_ehr_code_sql,
    dialect="standard",
    use_bqstorage_api=("BIGQUERY_STORAGE_API_ENABLED" in os.environ),
    progress_bar_type="tqdm_notebook")

ra_control_ehr_code.shape
```

```python
# last ehr record
ra_control_ehr_code = ra_control_ehr_code.loc[ra_control_ehr_code.groupby('person_id')['condition_start_datetime'].idxmax()]
ra_control_ehr_code.shape
```

```python
ra_control_ehr_code.head()
```

```python
# in control X is equivalent to last EHR follow-up time
ra_control = ra_control_person_df.merge(ra_control_ehr_code[['person_id', 'condition_start_datetime']], how = 'left')
ra_control.head()
```

```python
import numpy as np
ra_control['condition_start_datetime_ehr'] = ra_control['condition_start_datetime']
ra_control['ra_code_YN'] = np.where(ra_control['condition_start_datetime'].notna(), 0, None)
ra_control['ra_event_survey'] = np.where(ra_control['survey_datetime'].notna(), 0, None)
```

```python
ra_control.shape
```

```python
ra_control.head(20)
```

```python
ra_case.head()
```

```python
ra_df = pd.concat([ra_case, ra_control])
```

```python
ra_df.shape
```

```python
# Compute follow-up time (C)
ra_df['age_at_survey_event'] = ra_df.apply(
    lambda row: ((row['survey_datetime'] - row['date_of_birth']).days // 365) if pd.notna(row['survey_datetime']) else None, 
    axis=1
)

# Compute age at condition start time (X)
ra_df['age_at_condition_event'] = ra_df.apply(
    lambda row: ((row['condition_start_datetime'] - row['date_of_birth']).days // 365) if pd.notna(row['condition_start_datetime']) else None, 
    axis=1
)

# Compute last ehr follow-up time
ra_df['age_at_last_ehr'] = ra_df.apply(
    lambda row: ((row['condition_start_datetime_ehr'] - row['date_of_birth']).days // 365) if pd.notna(row['condition_start_datetime_ehr']) else None, 
    axis=1
)
```
