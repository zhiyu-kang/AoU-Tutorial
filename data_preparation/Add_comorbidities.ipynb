```python
# a participant may provide multiple answers at the same datetime.
RA_control_comorbidities_survey_df = RA_control_comorbidities_survey_df.drop_duplicates(subset='person_id', keep='first')
```

```python
# person with datetime answered "No" in RA survey, with NaT datetime didn't take the survey.
RA_control_comorbidities_person_df = RA_control_comorbidities_person_df.merge(RA_control_comorbidities_survey_df[["person_id", "survey_datetime"]], how = "left")
```

```python
ra_control_comorbidities_ehr_code_sql = """
    SELECT
        person_id,
        condition_concept_id,
        condition_start_datetime
    FROM
        `""" + os.environ["WORKSPACE_CDR"] + f""".condition_occurrence` 
    WHERE person_id in {tuple(ra_control_comorbidities_id)}
    """
ra_control_comorbidities_ehr_code = pd.read_gbq(
    ra_control_comorbidities_ehr_code_sql,
    dialect="standard",
    use_bqstorage_api=("BIGQUERY_STORAGE_API_ENABLED" in os.environ),
    progress_bar_type="tqdm_notebook")

ra_control_comorbidities_ehr_code.shape
```

```python
# last ehr record
ra_control_comorbidities_ehr_code = ra_control_comorbidities_ehr_code.loc[ra_control_comorbidities_ehr_code.groupby('person_id')['condition_start_datetime'].idxmax()]
ra_control_comorbidities_ehr_code.shape
```

```python
RA_control_comorbidities_person_df_sampled = RA_control_comorbidities_person_df[
    RA_control_comorbidities_person_df['person_id'].isin(ra_control_comorbidities_id)
]

print(f"Sampled DataFrame shape: {RA_control_comorbidities_person_df_sampled.shape}")
```

```python
# In control, age at condition start time (X) is equivalent to last EHR follow-up time (C)
ra_control_comorbidities = RA_control_comorbidities_person_df_sampled.merge(ra_control_comorbidities_ehr_code[['person_id', 'condition_start_datetime']], how = 'left')
```

```python
ra_control_comorbidities['condition_start_datetime_ehr'] = ra_control_comorbidities['condition_start_datetime']
ra_control_comorbidities['ra_code_YN'] = np.where(ra_control_comorbidities['condition_start_datetime'].notna(), 0, None)
ra_control_comorbidities['ra_event_survey'] = np.where(ra_control_comorbidities['survey_datetime'].notna(), 0, None)
```

```python
# Compute follow-up time (C)
ra_control_comorbidities['age_at_survey_event'] = ra_control_comorbidities.apply(
    lambda row: ((row['survey_datetime'] - row['date_of_birth']).days // 365) if pd.notna(row['survey_datetime']) else None, 
    axis=1
)

# Compute age at condition start time (X)
ra_control_comorbidities['age_at_condition_event'] = ra_control_comorbidities.apply(
    lambda row: ((row['condition_start_datetime'] - row['date_of_birth']).days // 365) if pd.notna(row['condition_start_datetime']) else None, 
    axis=1
)

# Compute last ehr follow-up time
ra_control_comorbidities['age_at_last_ehr'] = ra_control_comorbidities.apply(
    lambda row: ((row['condition_start_datetime_ehr'] - row['date_of_birth']).days // 365) if pd.notna(row['condition_start_datetime_ehr']) else None, 
    axis=1
)
```
