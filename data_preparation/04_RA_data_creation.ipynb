## Combine EHR and survey data with SNPs data

```python
# Load the dataset
ra_df = pd.read_csv('ra_demo_survey_code.csv')
```

```python
ra_df.head()
```

```python
ra_df.shape
```

```python
ra_case_snp = pd.read_csv('ra_case_selected_snp_matrix.csv')
ra_control_nonwhite1_snp = pd.read_csv('ra_control_nonwhite1_selected_snp_matrix.csv')
ra_control_nonwhite2_snp = pd.read_csv('ra_control_nonwhite2_selected_snp_matrix.csv')
ra_control_white1_snp = pd.read_csv('ra_control_white1_selected_snp_matrix.csv')
ra_control_white2_snp = pd.read_csv('ra_control_white2_selected_snp_matrix.csv')
```

```python
ra_snp = pd.concat([ra_case_snp, ra_control_nonwhite1_snp, ra_control_nonwhite2_snp, ra_control_white1_snp, ra_control_white2_snp])
```

```python
ra_snp.head()
```

```python
ra_snp.shape
```

```python
ra_df = ra_df.merge(ra_snp, how='inner')
```
